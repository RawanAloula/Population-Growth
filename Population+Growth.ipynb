### The Trouble with Tribbles

*This problem is part of Numerical Methods Course by Kyle Mandli.*

The number of tribbles (furry animals from Star Trek) is modeled by Spock to be exponential.  Spock predicts that in a 3 day period that the tribbles will go from an initial population number of $N_0 = 1$ to $N(3) = 1,771,561$ tribbles.  The assumptions that Spock made are the following:
 - The tribbles reproduce every 12 hours.
 - Tribbles reproduce asexually so that 1 tribble will produce 10 more tribbles leading to a total of 11 tribbles.


Can we write a function that confirms this estimation?

We use a population growth function 

$$
P=P_0~ e^{rt}
$$

where

$P_0= 1$

$P=1,771,561$

$t=6$


We know that population is observed over 3 days with 12 hours cycle. Instead of dividing time into days and hours we divid time into $\frac{72h}{12} = 6 $ production periods and then we solve the problem for r (the rate of change)

$e^{rt}= \frac{P}{P_0} $

$e^{r6}= \frac{1,771,561}{1} $

$ r = \frac{log(1,771,561)}{6} $ = $ 2.4$

Thus 

$$
P = e^{2.4t}
$$

````python
import numpy as np
import matplotlib.pyplot as plt

#plt.subplots_adjust(hspace=0.5)

#plt.suptitle('Tribbles Population Growth Over Time', fontsize=12)

x = np.linspace(0, 7)
y = np.exp(2.4*x)

#plt  x,y in linear scale
plt.subplot(211)
plt.plot(x, y, 'r')
plt.title('linear', fontsize=8)
plt.xlabel('Time (12 hours)',fontsize=9)
plt.ylabel('Population Growth',fontsize=9)

#plot with y in log scale
plt.subplot(212)
plt.semilogy(x,y, 'r')
plt.xlabel('Time (12 hours)',fontsize=9)
plt.ylabel('Population Growth',fontsize=9)
plt.title('Semi-log', fontsize=8)

plt.subplots_adjust(wspace=0.8, top=2)

plt.grid(True)
plt.show()
````

Each figure represent the population of that specific generation without adding up the contribution of previous generations. A good approach would be to visualize it as gif where each frame represent a generation as it adds up but that would be complicated as we need to incorporate the deaths as well. 

However, the above visualization was generated without using the population function 

$$
    N(x, y) = N_0 e^{-r^2/\sigma^2}
$$


where $N_0 = 1000$, $\sigma = 0.1$ and $r$ is the distance from the middle of our space ship. To use this function, we should first write the radius as 

$ r = (x^{2} + y^{2})^ \frac {1}{2} $

then generat random value for x  and  y to plot ( r , N(r) )

The code below takes this approach, but the visualization is not intuitive.
